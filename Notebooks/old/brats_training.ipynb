###2. Load Data 

```
#Loading data, must have shortcut to the shared macai_datasets folder in the base level of your google drive
train_dataset = wds.Dataset("macai_datasets/brats/train/brats_train.tar.gz")
eval_dataset = wds.Dataset("macai_datasets/brats/validation/brats_validation.tar.gz")
```

###3. Load Lightning Module

```
model = TumourSegmentation(learning_rate = 5e-5, train_collator=lambda x : col_img(x, use_zscore = True,pl_prob = 0.5),val_collator=lambda x : col_img(x,use_zscore = True,pl_prob = 0.0), batch_size=5, train_dataset=train_dataset, eval_dataset=eval_dataset)
```

###4. Train

```
wandb_logger = WandbLogger(project='macai',name='5e-5_zscore_0.5pl', offline = False)

#Training
trainer = pl.Trainer(
    accumulate_grad_batches = 1,
    gpus=1,
    max_epochs = 10,
    precision=16,
    check_val_every_n_epoch = 1,
    logger = wandb_logger,
    log_every_n_steps=10,      
    val_check_interval= 100,      
)
trainer.fit(model)
```
