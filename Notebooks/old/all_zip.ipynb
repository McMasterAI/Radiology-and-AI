```python
import gzip
import shutil
import os
```

```python
def do_zip(file):
    with open(file, 'rb') as f_in:
        with gzip.open(file+'.gz', 'wb') as f_out:
            shutil.copyfileobj(f_in, f_out)
    if os.path.exists(file):
        os.remove(file)
    else:
        print("The file does not exist")
```

```python
# aaa i should have threaded this, i forgot
def zip_everything(path):
    print(path)
    if not os.path.isdir(path): # If this iteration isn't a directory:
        do_zip(path)
    else:
        for new in os.listdir(path):
            new_path = os.path.join(path, new)
            zip_everything(new_path)
```
